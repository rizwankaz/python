```python
import matplotlib.pyplot as plt
import networkx as nx
import os

# Node positions
node_positions = {
    0: (461, 164),  1: (644, 415),  2: (461, 671),  3: (163, 575),
    4: (163, 260),  5: (419, 291),  6: (515, 415),  7: (419, 544),
    8: (271, 496),  9: (271, 339), 10: (433, 33),  11: (762, 483),
    12: (563, 759), 13: (33, 601), 14: (33, 246), 15: (561, 75),
    16: (762, 350), 17: (433, 802), 18: (109, 705), 19: (109, 139)
}

# Edges
edges = [
    (16, 1), (16, 17), (16, 15), (9, 6), (9, 6), (9, 7),
    (0, 10), (0, 5), (0, 15), (6, 9), (6, 1), (6, 8),
    (4, 9), (4, 14), (4, 19), (1, 16), (1, 6), (1, 11),
    (2, 12), (2, 17), (2, 7), (18, 3), (18, 17), (18, 19),
    (14, 4), (14, 15), (14, 13), (10, 0), (10, 11), (10, 19),
    (3, 18), (3, 8), (3, 13), (12, 2), (12, 11), (12, 13),
    (5, 0), (5, 8), (5, 7), (17, 16), (17, 2), (17, 18),
    (11, 1), (11, 10), (11, 12), (8, 6), (8, 3), (8, 5),
    (7, 9), (7, 2), (7, 5), (15, 16), (15, 0), (15, 14),
    (13, 14), (13, 3), (13, 12), (19, 4), (19, 18), (19, 10)
]

# Create the graph
G = nx.DiGraph()
G.add_edges_from(edges)

# Create output folder
output_folder = "chainweb"
os.makedirs(output_folder, exist_ok=True)

# Create and save one frame per highlighted node
for node_id in G.nodes():
    node_colors = ['#00ff00' if node == node_id else '#000000' for node in G.nodes()]
    
    plt.style.use('dark_background')
    fig, ax = plt.subplots(figsize=(10, 10))
    nx.draw_networkx_edges(G, pos=node_positions, edge_color='#00ff00', arrows=False, ax=ax)
    nx.draw_networkx_nodes(G, pos=node_positions,
                           node_color=node_colors,
                           edgecolors="#00ff00",
                           node_size=1000,
                           ax=ax)

    ax.set_axis_off()
    plt.tight_layout()
    
    frame_path = os.path.join(output_folder, f"green_{node_id:02d}.png")
    plt.savefig(frame_path, dpi=100)
    plt.close()

print(f"Saved frames to ./{output_folder}/")
```
